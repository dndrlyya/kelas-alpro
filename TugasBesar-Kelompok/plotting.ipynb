```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
#plt.style.use('style/sci.mplstyle')
import numpy as np

# Open and read the file gr.dos
b, x, y, a = np.loadtxt('energy.dat', unpack=True)
    
# Create figure object
plt.figure()
# Plot the DOS, in which the Fermi energy shifts to zero
plt.plot(x, color='red')
plt.plot(y, color='blue')
plt.plot(a, color='green')
# Add the x and y-axis labels
plt.xlabel('Timestamp')
plt.ylabel('EP, EK, EM (Joule)')
# Set the axis limits
plt.xlim(0, 20)
plt.ylim(0, 1000)
# Save the figure to the pdf file
plt.savefig('plotting.pdf', bbox_inches='tight')
```
